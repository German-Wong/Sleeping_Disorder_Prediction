```python
def describe_new(df): #function that adds range to the .describe() table
    df1 = df.describe() #stores  the original .describe() into df1
    df1.loc["range"] = df1.loc['max'] - df1.loc['min'] #creates another row for the ranges and calculates it by substracting max and min
    return df1

print(describe_new(df))
```

```python
df['Occupation'] = df['Occupation'].replace(['Nurse'], 0)
df['Occupation'] = df['Occupation'].replace(['Doctor'], 1)
df['Occupation'] = df['Occupation'].replace(['Engineer'], 2)
df['Occupation'] = df['Occupation'].replace(['Lawyer'], 3)
df['Occupation'] = df['Occupation'].replace(['Teacher'], 4)
df['Occupation'] = df['Occupation'].replace(['Accountant'], 5)
df['Occupation'] = df['Occupation'].replace(['Salesperson'], 6)
df['Occupation'] = df['Occupation'].replace(['Software Engineer'], 7)
df['Occupation'] = df['Occupation'].replace(['Scientist'], 8)
df['Occupation'] = df['Occupation'].replace(['Manager'], 9)

df['Occupation'].value_counts()
```

```python
#Replacing String Values to be represetned by Intergers
df['Gender'] = df['Gender'].replace(['Male'], 0)
df['Gender'] = df['Gender'].replace(['Female'], 1)
df['Gender'].value_counts()

df['BMI Category'] = df['BMI Category'].replace(['Normal'], 0)
df = df.replace(['Normal '], 0)
df['BMI Category'] = df['BMI Category'].replace(['Overweight'], 1)
df['BMI Category'] = df['BMI Category'].replace(['Obese'], 2)
df['BMI Category'].value_counts()
```
